### Dataset Description
The data used in this project is from the python package 'pybaseball'. This package is used for baseball data analysis and pulls data from Baseball Reference, Baseball Savant, and Fangraphs. The data pulled includes statcast data, pitching stats, batting stats, team stats and much more. The data can be viewed at a individual pitch level, a season level, or over a larger period of time. For a complete list of all functions, the 'pybaseball' documentation can be viewed on GitHub.

### Web Application Goal
The baseball savant web application replicated in this project is the 'Statcast Pitcher Velocity Distribution'. The goal of the web app is to visualize the frequency of all pitch types by pitch speed made by a certain pitcher in a certain year. The web app will make it easy to see how common / uncommon a certain pitch type is and how common / uncommon a certain pitch speed for that pitch type is. The web app will be able to show all pitch types or just the types manually chosen. The years that are able to be chosen range from 2008 - 2021 depending on which years the pitcher has played.

### Web Application Testing
To begin testing the web application, the python file must be opened and run in the terminal. The file first reads in the data which is a csv file containing the names, player ids, handedness, and years played for all pitchers who have played at least 50 innings. The file then splits the data into 2 datasets based on the handedness of the pitcher, one for left-handers, and one for right-handers. Next, the file creates a Dash web app and initializes 2 drop down menus, a graph, and a slider. There is one drop down menu for left-handed pitchers and one drop down menu for right-handed pitchers that contain the list of pitchers names. The graph updates to the distribution of the pitcher when chosen. The slider also updates when a pitcher is chosen but can range from 2008-2021. Lastly, the file specifies the callbacks everytime the web app is updated. The first callback disables the other drop down menu once a pitcher is chosen in a certain drop down menu. The second callback updates the slider everytime a pitcher is chosen. The last callback updates the graph once the pitcher and year is chosen. 

The csv data was retrieved by using the pitching_stats function to find all the names and ids of pitchers from 2008 to 2021 who had pitched at least 50 innings. The playerid_reverse_lookup function and the fangraph id was next used to find the pitchers names, mlbam id and the years played. The statcast_pitcher function and the apply method was used to create a new column that added the handedness of each pitcher. To update the graph, the statcast_pitcher function is used once again with the inputs of the selected year and the selected pitcher's id to retrieve the pitch type, pitch name, and release speed. A plotly histogram is then constructed with the data retreived.

As shown below, when the web application is first opened you will see a title of 'Statcast Pitcher Velocity Distribution', 2 drop down menus titled 'LHP' and 'RHP', a blank graph, and a slider wiht values from 2008-2021.

When you click on a drop down menu you will see all the options for that menu. The 'LHP' contains all the left handed pitchers and the 'RHP' contains all the right handed pitchers. Both list of pitchers are listed alphabetically and you can search for a pitcher as well.

Once a certain pitcher has been chosen, you will notice that the other drop down menu will be disabled. The menu will be enabled once you clear the current drop down menu. This is to allow only one pitcher to be chosen at a time. You will also notice that the slider will be updated to show the years played by the pitcher chosen. In this case, Shane Bieber has only played since 2018 so the slider will start at 2018. Since the data begins at 2008, if a player has been playing before 2008, the slider will still only start at 2008.

Once a pitcher has been chosen, you will have to manually select the year even if you want the year the slider is only on. After a couple of moments while app is gathering the necessary data, the graph will be updated with the pitch frequency. The title of the graph will be updated to show the selected pitcher and the selected year. The different pitches are all shown in different colors.

You can use the slider to change the year and the graph will update accordingly. Each update will take a couple of moments so you will have to be a little patient.

The plotly graph can also be manipulated to show only the pitch types you want. This can be updated by clicking on the pitch types in the legend to the right of the graph.
